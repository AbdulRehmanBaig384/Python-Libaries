# Data Cleaning and Missing values

```python
print(datapath.isnull().sum()) #in this code we search thee missing value in dataset and sum all the missing value and print it
```

```python
datapath.fillna(datapath.mean(),inplace=True) #this is method is too short for filling missing value to the collum means
# it give error because fillna method is always apply on numeric columns not text columns
```

```python
# Drop rows with missing values
datapath.dropna(inplace=True)
# this give no result because all dataset are proper not a single missing value
```

```python
# if we want to remove a columns which has missing values 
datapath.dropna(axis=1,inplace=True) # axis=1 indicates the column which has minimum value
# if we want to remove all column which has missing value then 
datapath.dropna(axis=1,how='all',inplace=True)
# if the column which has 50 percent missing value so remove it
datapath.dropna(thresh=len(datapath) * 0.5, axis=1, inplace=True)
```

```python
# this method is used if dataset has some missing value so it will put the datatype='number' in that missing column or numeric row  
datapath.fillna(datapath.select_dtypes(include=["number"]).mean(), inplace=True)
# if we want to remove all column which has missing value then 
datapath.fi
```

# Adding New Column

```python
# Calculate Total Price after Discount
datapath["Total_Price"] = datapath["Price"]*datapath["Quantity"]*(1 - datapath["Discount"])
datapath.head(10)
# Extract Month from Date
datapath["Date"] = pd.to_datetime(datapath["Date"])
datapath["Month"] = datapath["Date"].dt.month
```

```python
# in this cell we specified multiple rows 
row_multiple_specified=datapath.loc[[4,5,6,7,8]]
print(row_multiple_specified)
```

```python
# Locate Named Indexes
# Use the named index in the loc attribute to return the specified row(s)
specified_row=datapath.loc[1]
print(specified_row)
```
