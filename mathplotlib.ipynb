# Bar Plot

```python
import matplotlib.pyplot as plt
x=["Python","Java","C++","Dart"]
y=[80,40,30,20]
plt.xlabel("Programming labguages",fontsize=20)  # font size is used to increase the font size of the x label
plt.ylabel("Percetage of used")
plt.bar(x,y,width=0.4,color=["y","blue"],edgecolor="black",linewidth=2 ,linestyle="--",alpha=1), # align="edge",) align edge is used to write text just python on corner of the graph
plt.grid()
plt.show()
```

```python
x=["Python","Java","C++","Dart"]
y=[80,60,50,40]
z=[50,40,30,20]
plt.xlabel("Programming labguages",fontsize=20)  # font size is used to increase the font size of the x label
plt.ylabel("Percetage of used")
plt.bar(x,y,width=0.4,color=["y","blue"],edgecolor="black",label="Ubit"),
plt.bar(x,z,width=0.4,color=["r"],edgecolor="red",label="Ubit2")# align="edge",) align edge is used to write text just python on corner of the graph
#label is used to write something insde the table.to show label we should must right plt.legend
plt.legend()
plt.grid()
plt.show()
```

```python
# in order to show yand z side by side then we should followinf this code.frst we import numpy function then we used numpy build in function arrange.
import numpy as np
import matplotlib.pyplot as plt

x=["Python","Java","C++","Dart"]
y=[80,60,50,40]
z=[50,40,30,20]

width=0.2
p=np.arange(len(x))
p1=[j + width for j in p]

plt.xlabel("Programming labguages",fontsize=20)  # font size is used to increase the font size of the x label
plt.ylabel("Percetage of used")

plt.bar(p,y,width,color=["y","blue"],edgecolor="black",label="Ubit"),
plt.bar(p1,z,width,color=["r"],edgecolor="red",label="Ubit2")# align="edge",) align edge is used to write text just python on corner of the graph
#label is used to write something insde the table.to show label we should must right plt.legend

plt.xticks(p+width/2,x,rotation=20) #rick is used to show name variable like python,c++,java on axis and width/2 is used to place name of language like python java c++ in center of bar graph
# rotation is used to rotate x axis variable to a degree of number that is 20
plt.grid()
plt.legend()
plt.show()
```

#Scatter Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,4,1,6,5,4]
plt.scatter(x,y,color="r",s=100,edgecolor='black',cmap="viridis")
#marker="*") #s is used to define the size of dot or scatter plot
#marker is used to change colors
plt.colorbar()
plt.xlabel("Days")
plt.ylabel("Number")
plt.title("Scatter plot graph")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 4, 1, 6, 5, 4]
z = [3,4,5,6,7,8,9]
individual_color=[0,1,2,3,4,5,6] #it is necessary to write a color of random array and set into c variable for viridis color

plt.scatter(x, y, c=individual_color, s=200, edgecolor='black', cmap="viridis")
plt.scatter(x, z, c="r", s=200, edgecolor='black', )
plt.colorbar(label="Value of Y")  # Add a colorbar to show color scale
plt.xlabel("Days")
plt.ylabel("Number")
plt.title("Scatter Plot with Colormap")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating some random data between 0 and 100
no = np.random.randint(0, 100, 50)
mean_number=np.mean(no)

# range=(0, 100)
# The range parameter should be explicitly defined as range=(min, max).
# In your code, (0,100) is not assigned to any parameter, causing an error.

# bins=10 (or another value)
# plt.hist() requires bins to determine how many intervals to create. If you do not specify bins, it may not work as expected.

plt.hist(no, bins=10, range=(0, 200), edgecolor="blue", color="r")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.grid()
plt.axvline(mean_number,color="black",label="mean")
plt.legend()
plt.show()

```

# Pie Plot

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=["C","dart","C++","Java","Python"]
plt.pie(x,labels=y ,shadow=True)
plt.show()
```

```python
# in order to exclude a part of pie then we should do this.in this pie chart some properities we must given in dictionary form that inside a curly bracket {}
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=["C","dart","C++","Java","Python"]
ex=[0.5,0.0,0.0,0.0,0.0]
# wedgeprops={linewidth:value} is used to change the line width means border of the pie graph¶
# c=["r","black","blue","green","yellow"] color=c is used to change the color of the pie graph.
# in order to add peercentage in graph then we can used [autopct="%0.1f%%"]
# radius=0.2 is used to change the radius of pie chart to make it bgger and small
# labeldistance=0.8 is used to display text inside the pie chart
# startangle=90 is used to change the angle to a degree of,90,180,270,360 and many moreetc
# countercolock=False,True¶
# rotatelabel=False,True
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,radius=1,startangle=0,textprops={"fontsize":15},wedgeprops={"linewidth":6}),#"edgecolor":"black"})
plt.title("Programming languages")
plt.legend(loc=2)
```

# STEM PLOT

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,4,5,7,1,8]
# linefmt=":" is used to make line a dotted line
# markerfmt="r+") is used to make a circle to red with shape + signthat is r+,ro,r-
# bottom=1 is used to show graph start from x axis to 1
# basefmt="r" is used to change the base color to the red means the nichay wali lines
# orientation=
plt.stem(x,y,markerfmt="r",bottom=1,basefmt="r",label="Python",orientation="horizontal")
plt.legend()
plt.grid()
plt.legend()
plt.show()
```

# box plot

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
x1=[10,30,40,20,60,70,90]
y=[x,x1] #it is necessary to put both graph into single list 
# notch=True i used to change the shape of box to v shape
# vert=False is used to change the shape of box plot to a horizontal shape 
plt.boxplot(y,widths=0.3, notch=False,patch_artist=True)
plt.grid()
plt.legend()
plt.show()
```

```python
# in order to change the value of  to a name variables such as python so we used label in the form of list
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,120]
# notch=True i used to change the shape of box to v shape
# vert=False is used to change the shape of box plot to a horizontal shape 
# patch_artist=True is used to put color inside the whisper box plot
# showmeans=True is used t change the color of means means a small triangle of color green is formed on mean line
# small circle is formed ecause jab hum value a differencce ada ho jai uni philay 10,20,30,40,50 mai 10 ka difference hai jab hum yai likai 10,20,30,40,
# 50,120 likai tu jab 120 ai ga tu difference 70 ka ho gaya hai is liyai wo show karen ga a circel that the max range is 120
# sym="r" is used to change the color of the ball
# boxprops=dict(color="r") is used to change the border color of the box that is red
plt.boxplot(x,widths=0.3,labels=["Python"],patch_artist=True,showmeans=True,sym="r",boxprops=dict(color="r"))
plt.grid()
plt.legend()
plt.show()
```

# Stack plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[2,4,3,5,6]
area2=[3,1,4,7,5]
area3=[5,6,7,9,0]
l=["Area1","Area2","Area3"]
# baseline 
plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend()

plt.show()
```

# Step plot



```python
import matplotlib.pyplot as plt
x=[1,3,5,7,9]
y=[2,5,7,1,10]
# marker='o',ms=5 it is used to change make point of join and ms=5 is used to change the font soze of the marker 
plt.step(x,y,marker='o',ms=5,label="python") 
plt.title("Step plot ")
plt.xlabel(' x - axis')
plt.ylabel(' y - axis')
plt.legend()
plt.grid()
plt.show()
```

# Fill Between The Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.fill_between(x,y)
plt.plot(x,y,color='r')
plt.show()
```
